### 1) Momentos (Nota: 1.5/10)

A equação generalizada para o cálculo do momento $m_{pq}$ é dada por:

\begin{equation}
m_{pq} = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} x^p y^q \, f(x,y) \, dx dy, \end{equation}

onde $f(x,y)$ é a imagem, $x$ e $y$ suas respectivas coordenadas. Seguindo a teoria de momentos, as coordenadas do centro de massa normalizado de um objeto podem ser determinadas utilizando as seguintes equações:

\begin{equation}
  \bar{x} = \frac{m_{10}}{m_{00}}
  \;\;\;
  \bar{y} = \frac{m_{01}}{m_{00}}
\end{equation}

````python
## -- Seu código começa AQUI -- ##
# Definindo a função momentos
def momentos(img):
  '''
  Descrição: Calcula os momentos m00, m01, m10 e m11 de uma imagem.

  Entrada:
    - img: Imagem de entrada (uint8).
  
    Saída
    - moment_list: Lista contendo os 4 momentos calculados (float).
  '''
  m,n = img.shape # pega o tamanho da imagem

  moment_list = [] # cria a lista de momentos

  for p in range(2):
    for q in range(2):
      
      # Cálculo do momento 
      moment = 0 
      
      for x in range(m):
        for y in range(n):
          moment = moment + np.power(x,p)*np.power(y,q)*img[y,x]

      moment_list.append(moment) # Adicionando o momento atual à lista de momentos

  return moment_list

# Carregando as imagens
img1_tri  = cv.imread('triangulo.png', cv.IMREAD_GRAYSCALE)
img1_star = cv.imread('estrela.png', cv.IMREAD_GRAYSCALE)

# Aplicando a função momento nas imagens
moment_tri = momentos(img1_tri)
m00_tri = moment_tri[0]
m01_tri = moment_tri[1]
m10_tri = moment_tri[2]
m11_tri = moment_tri[3]

moment_star = momentos(img1_star)
m00_star = moment_star[0]
m01_star = moment_star[1]
m10_star = moment_star[2]
m11_star = moment_star[3]

# Calculando os centros de massa
xc_tri = int(m10_tri/m00_tri)
yc_tri = int(m01_tri/m00_tri)
print(f'O centro de massa do triangulo é {xc_tri},{yc_tri}')

xc_star = int(m10_star/m00_star)
yc_star = int(m01_star/m00_star)
print(f'O centro de massa da estrela é {xc_star},{yc_star}')

# Mostrando os resultados com o +
plt.title('Centro de massa do triangulo')
plt.imshow(img1_tri, cmap='gray')
plt.plot(xc_tri, yc_tri, '+')
plt.show()

plt.title('Centro de massa da estrela')
plt.imshow(img1_star, cmap='gray')
plt.plot(xc_star, yc_star, '+')
plt.show()

# Rotacionando a imagem do triangulo
img1_rot = cv.rotate(img1_tri, cv.ROTATE_180)

# Aplicando a função momento na imagem rotacionada
moment_rot = momentos(img1_rot)
m00_rot = moment_rot[0]
m01_rot = moment_rot[1]
m10_rot = moment_rot[2]
m11_rot = moment_rot[3]

# Calculando o centro de massa
xc_rot = int(m10_rot/m00_rot)
yc_rot = int(m01_rot/m00_rot)
print(f'O centro de massa do triangulo rotacionado é {xc_rot},{yc_rot}')

# Mostrando o resultado com o +
plt.title('Centro de massa do triangulo rotacionado')
plt.imshow(img1_rot, cmap='gray')
plt.plot(xc_rot, yc_rot, '+')
plt.show()
## -- Seu código termina AQUI -- ##

## -- COMENTÁRIOS -- ##
````

````python
def assinatura(img, ic, jc):
  '''
  Descrição: Cria assinatura para uma imagem.

  Entrada:
    - img: Imagem de entrada (uint8).
    - ic: Coordenada X do centro de massa normalizado (uint8).
    - jc: Coordenada Y do centro de massa normalizado (uint8).

  Saída:
    - rho: Coordenadas polar (rho) dos pontos [range=0:img.shape[0]*np.sqrt(2)/2].
    - theta: Coordenadas polar (theta) dos pontos [range=-180:180].
  '''
  # Sistema de coordenadas
  # .__> jj (y)
  # |
  # v
  # ii (x)

  # Teste de insanidade
  if img.ndim != 2:
    raise ValueError('Imagem deve ter 2 dimensões.')

  # Varre a imagem, determinando as coordenadas dos pontos brancos

  contours, hierarchy = cv.findContours(image=img, mode=cv.RETR_TREE, method=cv.CHAIN_APPROX_NONE)

  # Translada a origem do sistema de coordenadas para o centro de massa

  img_copy = np.zeros(img.shape)

  cv.drawContours(image=img_copy, contours=contours, contourIdx=1, color=(255, 255, 255), thickness=1, lineType=cv.LINE_AA)


  ii,jj = np.nonzero(img_copy)
  ii -= int(ic)
  jj -= int(jc)
  # Converte as novas coordenadas cartesianas para coordenadas polares
  rho = np.sqrt(ii**2 + jj**2)
  theta = np.rad2deg(np.arctan2(jj, ii))

  return rho, theta

## -- Seu código começa AQUI -- ##
# Carregando as imagens
img_c1 = cv.imread('circulo1.tif', cv.IMREAD_GRAYSCALE)
img_c2 = cv.imread('circulo2.tif', cv.IMREAD_GRAYSCALE)
img_q1 = cv.imread('quadrado1.tif', cv.IMREAD_GRAYSCALE)
img_q2 = cv.imread('quadrado2.tif', cv.IMREAD_GRAYSCALE)

# Calculando os centros de massa
moment_c1 = momentos(img_c1)
m00_c1 = moment_c1[0]
m01_c1 = moment_c1[1]
m10_c1 = moment_c1[2]
m11_c1 = moment_c1[3]

ic_c1 = (m10_c1/m00_c1).astype(np.uint8)
jc_c1 = (m01_c1/m00_c1).astype(np.uint8)

moment_c2 = momentos(img_c2)
m00_c2 = moment_c2[0]
m01_c2 = moment_c2[1]
m10_c2 = moment_c2[2]
m11_c2 = moment_c2[3]

ic_c2 = (m10_c2/m00_c2).astype(np.uint8)
jc_c2 = (m01_c2/m00_c2).astype(np.uint8)

moment_q1 = momentos(img_q1)
m00_q1 = moment_q1[0]
m01_q1 = moment_q1[1]
m10_q1 = moment_q1[2]
m11_q1 = moment_q1[3]

ic_q1 = (m10_q1/m00_q1).astype(np.uint8)
jc_q1 = (m01_q1/m00_q1).astype(np.uint8)

moment_q2 = momentos(img_q2)
m00_q2 = moment_q2[0]
m01_q2 = moment_q2[1]
m10_q2 = moment_q2[2]
m11_q2 = moment_q2[3]

ic_q2 = (m10_q2/m00_q2).astype(np.uint8)
jc_q2 = (m01_q2/m00_q2).astype(np.uint8)

# Calculando as assinaturas
rho_c1, theta_c1 = assinatura(img_c1, ic_c1, jc_c1)
rho_c2, theta_c2 = assinatura(img_c2, ic_c2, jc_c2)
rho_q1, theta_q1 = assinatura(img_q1, ic_q1, jc_q1)
rho_q2, theta_q2 = assinatura(img_q2, ic_q2, jc_q2)

# Mostrando as imagens

# Circulo 1
plt.subplot(1,2,1)
plt.title('Circulo 1')
plt.imshow(img_c1, cmap='gray')
plt.plot(ic_c1, jc_c1, '+')

plt.subplot(1,2,2)
plt.title('Assinatura circulo 1')
plt.plot(theta_c1,rho_c1,'.')
plt.axis([-180, 180, 0, 150])
plt.show()

# Circulo 2
plt.subplot(1,2,1)
plt.title('Circulo 2')
plt.imshow(img_c2, cmap='gray')
plt.plot(ic_c2, jc_c2, '+')

plt.subplot(1,2,2)
plt.title('Assinatura circulo 2')
plt.plot(theta_c2,rho_c2,'.')
plt.axis([-180, 180, 0, 150])
plt.show()

# Quadrado 1
plt.subplot(1,2,1)
plt.title('Quadrado 1')
plt.imshow(img_q1, cmap='gray')
plt.plot(ic_q1, jc_q1, '+')

plt.subplot(1,2,2)
plt.title('Assinatura quadrado 1')
plt.plot(theta_q1,rho_q1,'.')
plt.axis([-180, 180, 0, 150])
plt.show()

# Quadrado 2
plt.subplot(1,2,1)
plt.title('Quadrado 2')
plt.imshow(img_q2, cmap='gray')
plt.plot(ic_q2, jc_q2, '+')

plt.subplot(1,2,2)
plt.title('Assinatura quadrado 2')
plt.plot(theta_q2,rho_q2,'.')
plt.axis([-180, 180, 0, 150])
plt.show()

## -- Seu código termina AQUI -- ##

## -- COMENTÁRIOS -- ##
````

### Comentário

A assinatura varre a imagem do seu centro de massa até o ponto mais próximo onde encontra uma transição. Assim, os quadrados, que são idênticos, mas estão transladados entre si, possuem a mesma assinatura; isso implica que a assinatura é uma característica invariante à translação.

Quanto aos circulos, eles possuem uma asinatura muito parecida, por terem forma semelhante. Entretsnto, o circuio mais grosso possui um raio interno menor, de modo que a assinatura dele terá um nível menos que a do circulo mais fino.
